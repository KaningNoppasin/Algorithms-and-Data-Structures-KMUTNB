# Write a method insert(key, value, node) that inserts key and value into a binary search tree.

```python
class Node:
    def __init__(self, key=None, value=None):
        self.key = key
        self.value = value
        self.left = None
        self.right = None

class BST:
    def __init__(self):
        self.root = None

    def printTree(self):
        if self.root is not None:
            self._printTree(self.root)

    def _printTree(self, node, level=0):
        if node is not None:
            self._printTree(node.right, level + 1)
            print(' '*level*2 + '(' + str(node.key) + ',' + str(node.value) + ')')
            self._printTree(node.left, level + 1)

    def __setitem__(self, key, value):
        if self.root is None:
            self.root = Node(key, value)
        else:
            self.insert(key, value, self.root)

    def insert(self, key, value, node):
        if (key == node.key):
            node.key = key
            node.value = value
        elif (key < node.key):
            if node.left == None:
                node.left = Node(key, value)
            else:
                self.insert(key, value, node.left)
        elif (key > node.key):
            if node.right == None:
                node.right = Node(key, value)
            else:
                self.insert(key, value, node.right)
```

# Write a method search(key, node) that returns value associated with key in a binary search tree.

```python
class Node:
    def __init__(self, key=None, value=None):
        self.key = key
        self.value = value
        self.left = None
        self.right = None

class BST:
    def __init__(self):
        self.root = None

    def printTree(self):
        if self.root is not None:
            self._printTree(self.root)

    def _printTree(self, node, level=0):
        if node is not None:
            self._printTree(node.right, level + 1)
            print(' '*level*2 + '(' + str(node.key) + ',' + str(node.value) + ')')
            self._printTree(node.left, level + 1)

    def __getitem__(self, key):
        if self.root is not None:
            return self.search(key, self.root)
        else:
            return None 

    def search(self, key, node):
        if key == node.key:
            return node.value
        elif key < node.key:
            if node.left != None:
                return self.search(key, node.left)
        elif key > node.key:
            if node.right != None:
                return self.search(key,node.right)
        return None
```

# Write a method delete(key, node) that deletes element with key equal to key from a binary search tree.

```python
class Node:
    def __init__(self, key=None, value=None):
        self.key = key
        self.value = value
        self.left = None
        self.right = None

class BST:
    def __init__(self):
        self.root = None

    def printTree(self):
        if self.root is not None:
            self._printTree(self.root)

    def _printTree(self, node, level=0):
        if node is not None:
            self._printTree(node.right, level + 1)
            print(' '*level*2 + '(' + str(node.key) + ',' + str(node.value) + ')')
            self._printTree(node.left, level + 1)

    def maxValueNode(self, node):
        current = node
        while(current.right is not None):
            current = current.right
        return current

    def __delitem__(self, key):
        self.root = self.delete(key, self.root)

    def __setitem__(self, key, value):
        if self.root is None:
            self.root = Node(key, value)
        else:
            self.insert(key, value, self.root)

    def insert(self, key, value, node):
        if (key == node.key):
            node.key = key
            node.value = value
        elif (key < node.key):
            if node.left == None:
                node.left = Node(key, value)
            else:
                self.insert(key, value, node.left)
        elif (key > node.key):
            if node.right == None:
                node.right = Node(key, value)
            else:
                self.insert(key, value, node.right)

    def delete(self, key, node):
        if node == None:
            return None
        elif node.key > key:
            node.left = self.delete(key,node.left)
            return node
        elif node.key < key:
            node.right = self.delete(key,node.right)
            return node
        else:
            # Case 1:
            if node.left == None and node.right == None:
                node = None
                return node
            # Case 2:
            # Have Only Left
            elif node.left != None and node.right == None:
                node = node.left
                return node
            # Have Only Right
            elif node.left == None and node.right != None:
                node = node.right
                return node
            # Case 3:
            elif node.left != None and node.right != None:
                MaxNode = self.maxValueNode(node.left)
                node.left = self.delete(MaxNode.key, node.left)

                node.value = MaxNode.value
                node.key = MaxNode.key
                return node
```
